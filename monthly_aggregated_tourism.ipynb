```python
import pandas as pd
```

```python
# Read in data
tourism = pd.read_csv("JFK_LGA Arrivals from 2014-2024 - Sheet1.csv",skiprows=1,index_col=0)
tourism.head(30)
```

```python
## Create df just for JFK and LGA

jfk = tourism[0:12]
lga = tourism[16:28]

jfk.head()
lga.head()

```

```python
## FORMATTING jfk AND lga into 2 columns (1 for month-year, 1 for num of tourists thoughout month)

# Melt the DataFrame to long format
jfk = jfk.reset_index().rename(columns={'index': 'Month'})
jfk_melted = pd.melt(jfk,
                    id_vars=['Month'],
                    value_vars=[str(year) for year in range(2014, 2025)],
                    var_name='Year',
                    value_name='Number of Tourists (jfk)')

# Create the 'Month-Year' column
jfk_melted['Month-Year'] = jfk_melted['Month'] + '-' + jfk_melted['Year']

# Select and reorder the desired columns
jfk = jfk_melted[['Month-Year', 'Number of Tourists (jfk)']]
jfk['Number of Tourists (jfk)'] = jfk['Number of Tourists (jfk)'].str.replace(',', '', regex=False).astype(int)
jfk
```

```python
lga = lga.reset_index().rename(columns={'index': 'Month'})
lga_melted = pd.melt(lga,
                    id_vars=['Month'],
                    value_vars=[str(year) for year in range(2014, 2025)],
                    var_name='Year',
                    value_name='Number of Tourists (lga)')

# Create the 'Month-Year' column
lga_melted['Month-Year'] = lga_melted['Month'] + '-' + lga_melted['Year']

# Select and reorder the desired columns
lga = lga_melted[['Month-Year', 'Number of Tourists (lga)']]
lga['Number of Tourists (lga)'] = lga['Number of Tourists (lga)'].str.replace(',', '', regex=False).astype(int)
lga
```

```python
# ## Exporting lga dataset as csv
# lga.to_csv("lga_arrivals.csv", index=False)
```

```python
## Adding jfk and LGA to get total number of tourists


# Add the second columns
combined_col = pd.merge(jfk, lga, on='Month-Year', how='inner')
combined_col['Total_Tourists'] = combined_col['Number of Tourists (jfk)'] + combined_col['Number of Tourists (lga)']
combined = combined_col[['Month-Year', 'Total_Tourists']]
combined
```

```python
## creating master dataframe ##

monthly_tourism = pd.merge(jfk, lga, on='Month-Year', how='inner') #merging jfk and lga
monthly_tourism = pd.merge(monthly_tourism, combined, on='Month-Year', how='inner') # merging combined to jfk and lga
monthly_tourism
```

```python
# exporting monthly_tourism to csv
monthly_tourism.to_csv("monthly_tourism.csv", index=False)
```
