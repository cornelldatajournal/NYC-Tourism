```python
import pandas as pd
```

```python
## cleaning weather data
# reading in data
weather = pd.read_csv("CDJ Weather Data.csv")

# transposing dataframe
weather = weather.T
weather

#removing ANNUAL row and 2025 col
weather = weather.iloc[:13,:26]
weather

# drop NaN col #20
weather.drop(labels=20, axis=1, inplace=True)
weather

# convert years float --> int
weather.iloc[0,:] = weather.iloc[0,:].astype(int) #remove .0 
weather.iloc[0,:] = weather.iloc[0,:].astype(str) #convert to string for melt 
weather.columns = weather.iloc[0]
weather = weather[1:]
weather = weather.reset_index().rename(columns={'index': 'Month'})

weather
```

```python
## FORMATTING weather into 2 columns (1 for month-year, 1 for amount of precip for that month)

# Melt the DataFrame to long format
weather_melted = pd.melt(weather,
                    id_vars=['Month'],
                    value_vars=[str(year) for year in range(2000, 2025)], #converting years to strings
                    var_name='Year',
                    value_name='Precipitation')

# Create the 'Month-Year' column
weather_melted['Month-Year'] = weather_melted['Month'] + '-' + weather_melted['Year']

# Select and reorder the desired columns
weather = weather_melted[['Month-Year', 'Precipitation']]
weather
```

```python
# exporting monthly_precipitation to csv
weather.to_csv("monthly_precipitation.csv", index=False)
```
